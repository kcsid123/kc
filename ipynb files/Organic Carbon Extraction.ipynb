```python
for i in range(0,len(values)):
    values[i] = float(values[i])
    
    
values = np.array(values)
```

```python
lda_final = [0,0.05,0.2,0.3,0.6,0.9,1.2]

midpt = []
for i in range(0,len(lda_final)): 
    if i<len(lda_final)-1: 
        mid = (lda_final[i] + lda_final[i+1])/2
        
        midpt.append(mid)
```

```python
midpt = np.array(midpt)
values = np.array(values)

plt.plot(midpt, values, 'bo')

```

```python
from gekko import GEKKO

m = GEKKO()
m.x = m.Param(value = np.linspace(-0.05, 1.3))
m.y = m.Var()
m.cspline(m.x, m.y, midpt, values)
m.options.IMODE=2
m.solve(disp=False)


plt.plot(midpt, values, 'bo', label = 'data')
plt.plot(m.x, m.y, 'r--', label = 'cubic spline')
```

```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')

x = midpt
y = values

f = CubicSpline(x, y, bc_type='natural')
x_new = np.linspace(-0.05, 1.3)
y_new = f(x_new)
plt.figure(figsize = (10,8))
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro')
plt.title('Cubic Spline Interpolation')
plt.xlabel('Mid Point of Interval (meters)')
plt.ylabel('Average % of Sulfur')
plt.show()
```
