```python
tFile = txtFile[['LAB_METH_CODE', 'UNITS']]
tFile
```

```python
nullobsprop = master['ObservedProperty'].unique()

filecodes = tFile['LAB_METH_CODE'].unique()

matchingprops = []

checks = 0
for i in nullobsprop: 
    for j in filecodes:
        
        if i == j: 
            checks += 1
            matchingprops.append(i) 

print(len(filecodes), len(nullobsprop), end = '\n\n')
print('Number of matching properties are: ', checks, end = '\n')
```

```python
for i in matchingprops: 
    unit = tFile.loc[tFile['LAB_METH_CODE'] == i]['UNITS']
    
    master.loc[master['ObservedProperty'] == i, 'Units'] = unit
    
```
