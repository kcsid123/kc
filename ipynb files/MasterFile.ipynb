```python
#The Normal Bulk Density File Cannot be Used since there are too many NA Values for Lat/Long/Upper/Lower Depth 
bdv = bd2['LowerDepth']

# All values of Bulk Density in g/cm3
bdv = np.array(bdv)
bd2[bd2['Location_ID'] == '397_APSRU_243_1']

fig, axs = plt.subplots(1,1,figsize = (10,7), tight_layout = True)
axs.hist(bdv, bins = [0.15,0.3,0.6,0.9,1.2,1.5,1.8,2.1, 2.4, 2.7, 3.0])


for s in ['top', 'bottom', 'left', 'right']:
    axs.spines[s].set_visible(False)
 
axs.xaxis.set_ticks_position('none')
axs.yaxis.set_ticks_position('none')
   
# Add padding between axes and labels
axs.xaxis.set_tick_params(pad = 5)
axs.yaxis.set_tick_params(pad = 10)

axs.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.6)

N, bins, patches = axs.hist(bdv, bins = [0.15,0.3,0.6,0.9,1.2,1.5,1.8,2.1, 2.4, 2.7, 3.0] )
 
# Setting color
fracs = ((N**(1 / 5)) / N.max())
norm = colors.Normalize(fracs.min(), fracs.max())
 
for thisfrac, thispatch in zip(fracs, patches):
    color = plt.cm.viridis(norm(thisfrac))
    thispatch.set_facecolor(color)



plt.title('Bulk Density Occurence - Depth')
plt.xlabel('Bulk Density Intervals (0.3m = 1 interval)')
plt.ylabel('Number of Bulk Density Values below a certain depth')
plt.show()

# Gives us an idea of the methods that we could use for prediction. 
# Lower BD values might be better predicted with a different method, as compared to predict higher BD values. 
```

```python
cat1 = []
cat2 = []
cat3 = []
cat4 = []
cat5 = []
catl = []



bd2a = bd2

for row in bd2a.index: 
    if(bd2a['LowerDepth'][row] <= 0.3): 
        cat1.append(bd2a['Value'][row])
       
        
    elif (bd2a['LowerDepth'][row] <= 0.6): 
        cat2.append(bd2a['Value'][row])
        
        
    elif (bd2a['LowerDepth'][row] <= 0.9): 
        cat3.append(bd2a['Value'][row])
       
        
    elif (bd2a['LowerDepth'][row] <= 1.2): 
        cat4.append(bd2a['Value'][row])
        
        
    elif (bd2a['LowerDepth'][row] <= 1.5): 
        cat5.append(bd2a['Value'][row])
        
        
    else: 
        catl.append(bd2a['Value'][row])



lens = [len(cat1), len(cat2), len(cat3), len(cat4), len(cat5), len(catl)]
```

```python
fig1, axs1 = plt.subplots(1,1,figsize = (10,7), tight_layout = True)
axs1.hist(lens, bins = [0,0.3,0.6,0.9,1.2,1.5,1.8])


for s in ['top', 'bottom', 'left', 'right']:
    axs1.spines[s].set_visible(False)
 
axs1.xaxis.set_ticks_position('none')
axs1.yaxis.set_ticks_position('none')
   
# Add padding between axes and labels
axs1.xaxis.set_tick_params(pad = 5)
axs1.yaxis.set_tick_params(pad = 10)

axs1.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.6)

N, bins, patches = axs1.hist(lens, bins = [0,0.3,0.6,0.9,1.2,1.5,1.8] )
 
# Setting color
fracs = ((N**(1 / 5)) / N.max())
norm = colors.Normalize(fracs.min(), fracs.max())
 
for thisfrac, thispatch in zip(fracs, patches):
    color = plt.cm.viridis(norm(thisfrac))
    thispatch.set_facecolor(color)



plt.title('Consolidated Bulk Density - Values')
plt.xlabel('Intervals (0.3 g/cm3)')
plt.ylabel('Number of Bulk Density Values in bins(g/cm3)')
plt.show()
```
